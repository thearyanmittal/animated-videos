```
%%manim Thing22
from manimlib.scene.scene import Scene
from manimlib.mobject.geometry import Circle
from manimlib.animation.creation import ShowCreation
from manimlib.imports import *

class Thing22(Scene):
  def construct(self):
    goal_label = TextMobject("The equation we want to get:")
    goal_label.set_color_by_gradient(RED_D, YELLOW, BLUE_D)
    goal = TextMobject("3" + "$\t{C}$" + " + " + "4" + "$\t{H}_{\t{2}}$" + " = " + "$\t{C}_{\t{3}}$" + "$\t{H}_{\t{8}}$")
    goal.set_color(GREEN_SCREEN)

    given_label = TextMobject("The equations we are given:")
    given_label.set_color_by_gradient(RED_D, YELLOW, BLUE_D)

    first_line = TextMobject("$\t{C}_{\t{3}}$" + "$\t{H}_{\t{8}}$" + " + " + "5" + "$\t{O}_{\t{2}}$" + " = " + "3" + "$\t{CO}_{\t{2}}$" + " + " + "4" + "$\t{H}_{\t{2}}$" + "$\t{O}$" + " " + "$\Delta$" + "H = -2219.2 kJ")
    first_line.set_color(YELLOW)

    second_line = TextMobject("$\t{C}$" + " + " + "$\t{O}_{\t{2}}$" + " = " + "$\t{CO}_{\t{2}}$" + " " + "$\Delta$" + "H = -393.5 kJ")
    second_line.set_color(YELLOW)
    second_line.next_to(first_line, DOWN)

    third_line = TextMobject("$\t{H}_{\t{2}}$" + " + " + "$\t{1/2}$" + "$\t{O}_{\t{2}}$" + " = " + "$\t{H}_{\t{2}}$" + "$\t{O}$" + " " + "$\Delta$" + "H = -285.8 kJ")
    third_line.set_color(YELLOW)
    third_line.next_to(second_line, DOWN)

    og_first_line = TextMobject("$\t{C}_{\t{3}}$" + "$\t{H}_{\t{8}}$" + " + " + "5" + "$\t{O}_{\t{2}}$" + " = " + "3" + "$\t{CO}_{\t{2}}$" + " + " + "4" + "$\t{H}_{\t{2}}$" + "$\t{O}$" + " " + "$\Delta$" + "H = -2219.2 kJ")
    og_first_line.set_color(YELLOW)
    og_second_line = TextMobject("$\t{C}$" + " + " + "$\t{O}_{\t{2}}$" + " = " + "$\t{CO}_{\t{2}}$" + " " + "$\Delta$" + "H = -393.5 kJ")
    og_second_line.set_color(YELLOW)
    og_second_line.next_to(og_first_line, DOWN)
    og_third_line = TextMobject("$\t{H}_{\t{2}}$" + " + " + "$\t{1/2}$" + "$\t{O}_{\t{2}}$" + " = " + "$\t{H}_{\t{2}}$" + "$\t{O}$" + " " + "$\Delta$" + "H = -285.8 kJ")
    og_third_line.set_color(YELLOW)
    og_third_line.next_to(og_second_line, DOWN)

    first_line_ex1 = TextMobject("We'll get to the heats of reaction later...")
    first_line_ex1.set_color_by_gradient(RED_D, YELLOW, BLUE_D)
    first_line_ex1.to_edge(DOWN)
    first_line_ex2 = TextMobject("Flip the first equation to get " + "$\t{C}_{\t{3}}$" + "$\t{H}_{\t{8}}$" + " on the products side")
    first_line_ex2.set_color_by_gradient(RED_D, YELLOW, BLUE_D)
    first_line_ex2.to_edge(DOWN)

    second_line_ex1 = TextMobject("Notice the 3 coefficient in front of the " + "$\t{CO}_{\t{2}}$" + " in the first equation")
    second_line_ex1.scale(0.85)
    second_line_ex1.set_color_by_gradient(RED_D, YELLOW, BLUE_D)
    second_line_ex1.to_edge(DOWN)
    second_line_ex2 = TextMobject("Multiply the second equation by 3 so we can cancel the " + "$\t{CO}_{\t{2}}$")
    second_line_ex2.set_color_by_gradient(RED_D, YELLOW, BLUE_D)
    second_line_ex2.to_edge(DOWN)
    second_line_ex3 = TextMobject("Notice that we now have a 3 in front of " + "$\t{C}$" + ", like the green equation")
    second_line_ex3.set_color_by_gradient(RED_D, YELLOW, BLUE_D)
    second_line_ex3.to_edge(DOWN)
    second_line_ex3.scale(0.85)

    third_line_ex1 = TextMobject("Notice the 4 coefficient in front of the " + "$\t{H}_{\t{2}}$" + "$\t{O}$" + " in the first red and blue equation")
    third_line_ex1.scale(0.8)
    third_line_ex1.set_color_by_gradient(RED_D, YELLOW, BLUE_D)
    third_line_ex1.to_edge(DOWN)
    third_line_ex2 = TextMobject("Multiply the third equation by 4 so we can cancel the " + "$\t{H}_{\t{2}}$" + "$\t{O}$")
    third_line_ex2.set_color_by_gradient(RED_D, YELLOW, BLUE_D)
    third_line_ex2.to_edge(DOWN)
    third_line_ex3 = TextMobject("Notice we now have a 4 in front of the " + "$\t{H}_{\t{2}}$" + ", like in the green equation")
    third_line_ex3.set_color_by_gradient(RED_D, YELLOW, BLUE_D)
    third_line_ex3.to_edge(DOWN)
    third_line_ex3.scale(0.85)

    new_first_line = TextMobject("3" + "$\t{CO}_{\t{2}}$" + " + " + "4" + "$\t{H}_{\t{2}}$" + "$\t{O}$")
    new_first_line.set_color(RED)
    new_first_line2= TextMobject(" = " + "$\t{C}_{\t{3}}$" + "$\t{H}_{\t{8}}$")
    new_first_line2.set_color(BLUE_D)
    new_first_line2.next_to(new_first_line, RIGHT)
    new_first_line3 = TextMobject(" + " + "5" + "$\t{O}_{\t{2}}$")
    new_first_line3.set_color(RED)
    new_first_line3.next_to(new_first_line2, RIGHT)

    new_second_line = TextMobject("3" + "$\t{C}$" + " + ")
    new_second_line.set_color(BLUE_D)
    new_second_line.next_to(new_first_line, DOWN)
    new_second_line2 = TextMobject("3" + "$\t{O}_{\t{2}}$" + " = " + "3" + "$\t{CO}_{\t{2}}$")
    new_second_line2.set_color(RED)
    new_second_line2.next_to(new_second_line, RIGHT)

    new_third_line = TextMobject("4" + "$\t{H}_{\t{2}}$" + " + ")
    new_third_line.set_color(BLUE_D)
    new_third_line.next_to(new_second_line, DOWN)
    new_third_line2 = TextMobject("2" + "$\t{O}_{\t{2}}$" + " = " + "4" + "$\t{H}_{\t{2}}$" + "$\t{O}$")
    new_third_line2.set_color(RED)
    new_third_line2.next_to(new_third_line, RIGHT)

    newest_first_line = TextMobject(" = ")
    newest_first_line.set_color(ORANGE)
    newest_first_line2 = TextMobject("$\t{C}_{\t{3}}$" + "$\t{H}_{\t{8}}$")
    newest_first_line2.set_color(ORANGE)
    newest_first_line2.next_to(newest_first_line, RIGHT)
    
    newest_second_line = TextMobject(" = ")
    newest_second_line.set_color(ORANGE)
    newest_second_line.next_to(newest_first_line, DOWN)
    newest_second_line2 = TextMobject("3" + "$\t{C}$")
    newest_second_line2.set_color(ORANGE)
    newest_second_line2.next_to(newest_second_line, LEFT)

    newest_third_line = TextMobject(" = ")
    newest_third_line.set_color(ORANGE)
    newest_third_line.next_to(newest_second_line, DOWN)
    newest_third_line.shift(0.1*DOWN)
    newest_third_line2 = TextMobject("4" + "$\t{H}_{\t{2}}$")
    newest_third_line2.set_color(ORANGE)
    newest_third_line2.next_to(newest_third_line, LEFT)

    cancel_ex1 = TextMobject("Cancel out terms that are the same and on opposite sides of the equals sign.")
    cancel_ex1.scale(0.85)
    cancel_ex1.set_color_by_gradient(RED_D, YELLOW, BLUE_D)
    cancel_ex1.to_edge(DOWN)
    cancel_ex2 = TextMobject("3" + "$\t{CO}_{\t{2}}$")
    cancel_ex2.set_color(RED)
    cancel_ex2.to_edge(DOWN)
    cancel_ex3 = TextMobject("4" + "$\t{H}_{\t{2}}$" + "$\t{O}$")
    cancel_ex3.set_color(RED)
    cancel_ex3.to_edge(DOWN)
    cancel_ex4 = TextMobject("3" + "$\t{O}_{\t{2}}$" + " + " + "2" + "$\t{O}_{\t{2}}$" + " = " + "5" + "$\t{O}_{\t{2}}$")
    cancel_ex4.set_color(RED)
    cancel_ex4.to_edge(DOWN)

    finale_ex1 = TextMobject("We have the ingredients for the final equation!")
    finale_ex1.set_color_by_gradient(RED_D, YELLOW, BLUE_D)
    finale_ex1.to_edge(DOWN)
    finale_ex2 = TextMobject("Carbon and Hydrogen on the left of the equals sign add to form Propane on the right")
    finale_ex2.scale(0.65)
    finale_ex2.set_color_by_gradient(RED_D, YELLOW, BLUE_D)
    finale_ex2.to_edge(DOWN)

    heat1 = TextMobject("$\Delta$" + "H = 2219.2 kJ")
    heat1.set_color(PURPLE)
    heat2 = TextMobject("$\Delta$" + "H = -1180.5 kJ")
    heat2.set_color(PURPLE)
    heat2.next_to(heat1, DOWN)
    heat3 = TextMobject("$\Delta$" + "H = -1143.2 kJ")
    heat3.set_color(PURPLE)
    heat3.next_to(heat2, DOWN)
    goal_heat = TextMobject("$\Delta$" + "H = -104.5 kJ")
    goal_heat.set_color(GREEN_SCREEN)
    goal_heat.next_to(goal, RIGHT)

    heat_ex1 = TextMobject("Remember those heats of reaction? Let's pull 'em out again")
    heat_ex1.set_color_by_gradient(RED_D, YELLOW, BLUE_D)
    heat_ex1.to_edge(DOWN)
    heat_ex2 = TextMobject("Scale the heats according to what we did to their respective equations")
    heat_ex2.scale(0.9)
    heat_ex2.set_color_by_gradient(RED_D, YELLOW, BLUE_D)
    heat_ex2.to_edge(DOWN)
    heat_ex3 = TextMobject("Flip the sign of -2219.2 kJ since we flipped its equation")
    heat_ex3.set_color_by_gradient(RED_D, YELLOW, BLUE_D)
    heat_ex3.to_edge(DOWN)
    heat_ex4 = TextMobject("Multiply -393.5 kJ by 3 since we multiplied its equation by 3")
    heat_ex4.set_color_by_gradient(RED_D, YELLOW, BLUE_D)
    heat_ex4.to_edge(DOWN)
    heat_ex5 = TextMobject("Multiply -285.8 kJ by 4 since we multiplied its equation by 4")
    heat_ex5.set_color_by_gradient(RED_D, YELLOW, BLUE_D)
    heat_ex5.to_edge(DOWN)
    heat_ex6 = TextMobject("Now, add them together to get " + "$\Delta$" + "H for the final equation")
    heat_ex5.scale(0.9)
    heat_ex6.set_color_by_gradient(RED_D, YELLOW, BLUE_D)
    heat_ex6.to_edge(DOWN)

    self.play(Write(goal_label))
    self.wait(2)
    self.play(FadeOut(goal_label))
    self.play(Write(goal))
    self.play(ApplyMethod(goal.shift, 3.1*UP))
    self.wait(2)

    self.play(Write(given_label))
    self.wait(2)
    self.play(FadeOut(given_label))

    self.play(Write(first_line))
    self.wait(1)
    self.play(Write(second_line))
    self.wait(1)
    self.play(Write(third_line))
    self.wait(2)

    self.play(ApplyMethod(first_line.shift, 2*UP), ApplyMethod(second_line.shift, 2*UP), ApplyMethod(third_line.shift, 2*UP))
    self.wait(2)

    self.play(Write(first_line_ex1))
    self.wait(1)
    self.play(Transform(first_line_ex1, first_line_ex2))
    self.wait(1)
    self.play(Transform(first_line, new_first_line))
    self.play(Write(new_first_line2))
    self.play(Write(new_first_line3))
    self.play(ApplyMethod(first_line.shift, 2*LEFT), ApplyMethod(new_first_line.shift, 2*LEFT), ApplyMethod(new_first_line2.shift, 2*LEFT), ApplyMethod(new_first_line3.shift, 2*LEFT))
    self.wait(1)
    self.play(FadeOut(first_line_ex2), FadeOut(first_line_ex1))
    self.wait(1)

    self.play(Write(second_line_ex1))
    self.wait(3)
    self.play(Transform(second_line_ex1, second_line_ex2))
    self.wait(2)
    self.play(Transform(second_line, new_second_line))
    self.play(Write(new_second_line2))
    self.play(ApplyMethod(second_line.shift, 2*LEFT), ApplyMethod(new_second_line.shift, 2*LEFT), ApplyMethod(new_second_line2.shift, 2*LEFT))
    self.wait(1)
    self.play(Transform(second_line_ex1, second_line_ex3), Transform(second_line_ex2, second_line_ex3))
    self.wait(3)
    self.play(FadeOut(second_line_ex3), FadeOut(second_line_ex2), FadeOut(second_line_ex1))
    self.wait(1)

    self.play(Write(third_line_ex1))
    self.wait(3)
    self.play(Transform(third_line_ex1, third_line_ex2))
    self.wait(2)
    self.play(Transform(third_line, new_third_line))
    self.play(Write(new_third_line2))
    self.play(ApplyMethod(third_line.shift, 2*LEFT), ApplyMethod(new_third_line2.shift, 2*LEFT))
    self.wait(1)
    self.play(Transform(third_line_ex2, third_line_ex3), Transform(third_line_ex1, third_line_ex3))
    self.wait(3)
    self.play(FadeOut(third_line_ex3), FadeOut(third_line_ex2), FadeOut(third_line_ex1))
    self.wait(1)
    
    self.play(Write(cancel_ex1))
    self.wait(3)
    self.play(Transform(cancel_ex1, cancel_ex2))
    self.wait(1)
    self.play(Transform(cancel_ex1, cancel_ex3), Transform(cancel_ex2, cancel_ex3))
    self.wait(1)
    self.play(Transform(cancel_ex1, cancel_ex4), Transform(cancel_ex2, cancel_ex4), Transform(cancel_ex3, cancel_ex4))
    self.wait(0.5)
    self.play(FadeOut(cancel_ex1), FadeOut(cancel_ex2), FadeOut(cancel_ex3), FadeOut(cancel_ex4))
    self.play(FadeOut(first_line), FadeOut(second_line), FadeOut(third_line), FadeOut(new_first_line2), FadeOut(new_first_line3), FadeOut(new_second_line2), FadeOut(new_third_line2), 
              FadeOut(new_first_line), FadeOut(new_second_line), FadeOut(new_third_line))
    self.wait(1)
    self.play(Write(newest_first_line), Write(newest_first_line2))
    self.play(Write(newest_second_line), Write(newest_second_line2))
    self.play(Write(newest_third_line), Write(newest_third_line2))
    self.wait(3)

    self.play(Write(finale_ex1))
    self.wait(1)
    self.play(ApplyMethod(newest_first_line.shift, UP), ApplyMethod(newest_first_line2.shift, UP), ApplyMethod(newest_second_line.shift, UP), 
              ApplyMethod(newest_second_line2.shift, UP), ApplyMethod(newest_third_line.shift, UP), ApplyMethod(newest_third_line2.shift, UP))
    self.play(Transform(finale_ex1, finale_ex2))
    self.wait(4)
    self.play(FadeOut(finale_ex1), FadeOut(finale_ex2))
    self.play(ApplyMethod(goal.shift, 4*DOWN))
    self.wait(2)

    self.play(ApplyMethod(goal.shift, 4*UP))
    self.play(FadeOut(newest_first_line), FadeOut(newest_first_line2), FadeOut(newest_second_line), FadeOut(newest_second_line2), FadeOut(newest_third_line), FadeOut(newest_third_line2))
    self.wait(1)
    self.play(Write(heat_ex1))
    self.wait(2)
    self.play(Write(og_first_line), Write(og_second_line), Write(og_third_line))
    self.play(ApplyMethod(og_first_line.shift, 2*UP), ApplyMethod(og_second_line.shift, 2*UP), ApplyMethod(og_third_line.shift, 2*UP))
    self.wait(1)
    self.play(Transform(heat_ex1, heat_ex2))
    self.wait(2)
    self.play(Transform(heat_ex2, heat_ex3), Transform(heat_ex1, heat_ex3))
    self.wait(2)
    self.play(Transform(og_first_line, heat1))
    self.play(FadeOut(heat_ex1), FadeOut(heat_ex2), FadeOut(heat_ex3))
    self.wait(1)
    self.play(Write(heat_ex4))
    self.wait(2)
    self.play(Transform(og_second_line, heat2))
    self.wait(1)
    self.play(Transform(heat_ex4, heat_ex5))
    self.wait(2)
    self.play(Transform(og_third_line, heat3))
    self.wait(1)
    self.play(ApplyMethod(og_first_line.shift, 2*UP), ApplyMethod(og_second_line.shift, 2*UP), ApplyMethod(og_third_line.shift, 2*UP), ApplyMethod(heat1.shift, 2*UP), ApplyMethod(heat2.shift, 2*UP), 
              ApplyMethod(heat3.shift, 2*UP))
    self.play(FadeOut(heat_ex4), FadeOut(heat_ex5))
    self.play(Write(heat_ex6))
    self.wait(1.5)
    self.play(ApplyMethod(goal.shift, 3.1*DOWN))
    self.play(Write(goal_heat))
    self.play(ApplyMethod(goal.shift, 2*LEFT), ApplyMethod(goal_heat.shift, 2*LEFT), ApplyMethod(heat1.shift, 2*RIGHT), ApplyMethod(heat2.shift, 2*RIGHT), ApplyMethod(heat3.shift, 2*RIGHT), 
              ApplyMethod(og_first_line.shift, 2*RIGHT), ApplyMethod(og_second_line.shift, 2*RIGHT), ApplyMethod(og_third_line.shift, 2*RIGHT))
    self.wait(3)
```
