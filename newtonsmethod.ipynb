```
%%manim NewtonPART1
from manimlib.scene.scene import *
from manimlib.mobject.geometry import *
from manimlib.animation.creation import *
from manimlib.imports import *

class NewtonPART1(GraphScene):


  CONFIG = {
        "x_min": -7.12,
        "x_max": 7.12,
        "x_axis_width": 14.3,
        "x_tick_frequency": 15,
        "x_leftmost_tick": 0,  # Change if different from x_min
        "x_labeled_nums": None,
        "x_axis_label": None,
        "y_min": -4.01,
        "y_max": 4.01,
        "y_axis_height": 8.0,
        "y_tick_frequency": 15,
        "y_bottom_tick": 0,  # Change if different from y_min
        "y_labeled_nums": None,
        "y_axis_label": None,
        "axes_color": WHITE,
        "graph_origin": ORIGIN,
        "exclude_zero_label": True,
        "default_graph_colors": [BLUE, GREEN, YELLOW],
        "default_derivative_color": GREEN,
        "default_input_color": YELLOW,
        "default_riemann_start_color": BLUE,
        "default_riemann_end_color": GREEN,
        "area_opacity": 0.8,
        "num_rects": 50,
    }

  def graph_function(self):
    
    def func(x):
      return 2*x**3 + x**2 - x + 1

    graph = self.get_graph(func, GREEN, x_min=-4, x_max=4)
    label = self.get_graph_label(graph, label='f(x) = 2x^3 + x^2 - x + 1', color=GREEN)
    point = Dot(point=np.array([-1.234, 0.0, 0.0]), color=GREEN_SCREEN)
    point_arrow = Arrow(start=np.array([-3.5, 2.0, 0]), end=np.array([-1.234, 0.05, 0.0]), color=MAROON)
    point_label = TextMobject('Can we approximate this zero?', color=MAROON).move_to([-3.5, 2.25, 0.0])
    point_label2 = TextMobject('Yes! With Newton\'s Method!', color=RED).move_to([-3.5, 2.25, 0.0])
    point_label3 = TextMobject('First, let\'s focus on just one part of the graph', color=YELLOW).move_to([0, -3.0, 0.0])

    # Display graph
    self.play(ShowCreation(graph, run_time=5.0))
    self.play(Write(label))
    self.wait(2)

    self.play(ShowCreation(point))
    self.wait(1)
    self.play(Write(point_label), ShowCreation(point_arrow))
    self.wait(1.5)
    self.play(ReplacementTransform(point_label, point_label2), Uncreate(point_arrow))
    self.wait(1.5)
    self.play(ReplacementTransform(point_label2, point_label3))
    self.wait(2)


  def construct(self):
    x_label = TextMobject('$x$').move_to([6.75, 0.25, 0])
    y_label = TextMobject('$y$').move_to([0.25, 3.75, 0])
    origin_label = TextMobject('$O$').move_to([-0.25, -0.25, 0.0])
    plane = NumberPlane()

    self.play(ShowCreation(plane))
    self.setup_axes(animate=True)
    self.play(Write(x_label), Write(y_label), Write(origin_label))    

    self.graph_function()
    self.play(*[Uncreate(mob) for mob in self.mobjects], run_time=5.0)
```

```
%%manim NewtonPART2
from manimlib.scene.scene import *
from manimlib.mobject.geometry import *
from manimlib.animation.creation import *
from manimlib.imports import *

class NewNumberPlane(NumberPlane):
  

  CONFIG = {
        "axis_config": {
            "stroke_color": BLUE_D,
            "stroke_width": 2,
            "include_ticks": False,
            "include_tip": False,
            "line_to_number_buff": SMALL_BUFF,
            "label_direction": DR,
            "number_scale_val": 0.5,
        },
        "y_axis_config": {
            "label_direction": DR,
        },
        "background_line_style": {
            "stroke_color": BLUE_D,
            "stroke_width": 2,
            "stroke_opacity": 1,
        },
        # Defaults to a faded version of line_config
        "faded_line_style": None,
        "x_line_frequency": 1,
        "y_line_frequency": 1,
        "faded_line_ratio": 1,
        "make_smooth_after_applying_functions": True,
    }

class NewtonPART2(GraphScene):


  CONFIG = {
        "x_min": -2.20,
        "x_max": 1.34,
        "x_axis_width": 18,
        "x_tick_frequency": 25,
        "x_leftmost_tick": 0,  # Change if different from x_min
        "x_labeled_nums": [-2, -1],
        "x_axis_label": None,
        "y_min": -3.67,
        "y_max": 2.34,
        "y_axis_height": 8.5,
        "y_tick_frequency": 15,
        "y_bottom_tick": 0,  # Change if different from y_min
        "y_labeled_nums": [-3, -2, -1, 1, 2],
        "y_axis_label": None,
        "axes_color": WHITE,
        "graph_origin": np.array([4.0, 1.0, 0.0]),
        "exclude_zero_label": False,
        "default_graph_colors": [BLUE, GREEN, YELLOW],
        "default_derivative_color": GREEN,
        "default_input_color": YELLOW,
        "default_riemann_start_color": BLUE,
        "default_riemann_end_color": GREEN,
        "area_opacity": 0.8,
        "num_rects": 50,
    }



  def tangents(self):

    def func(x):
      return 2*x**3 + x**2 - x + 1

    graph = self.get_graph(func, GREEN, x_min=-5.0, x_max=5.0)
    point = Dot(color=GREEN_SCREEN).move_to(self.coords_to_point(-1.234, 0.0))

    self.play(ShowCreation(graph, rate_func=linear, run_time=5.0))

    self.play(ShowCreation(point))
    self.wait(2)
    
    x0_point = Dot(color=BLUE_D).move_to(self.coords_to_point(-0.9, 0.0))
    x0_point_small = Dot(color=BLUE_D, radius=0.05).move_to(self.coords_to_point(-0.9, 0.0)) 
    x0_point_label = TextMobject('$x_0$', color=BLUE).scale(0.75).move_to(self.coords_to_point(-0.9,-0.2))
    x0_point_to_graph = self.get_vertical_line_to_graph(-0.9, graph).set_color(PURPLE)

    tang1 = TangentLine(graph, 0.41, length=20)
    tang1.set_color(YELLOW)
    tang1_label = TextMobject('$y-1.25=2.06(x+0.90)$', color=YELLOW).move_to([0.0, -2.0, 0.0])
    tang1_point = Dot(color=YELLOW_E).move_to(self.coords_to_point(-1.508, 0.0))
    tang1_point_small = Dot(color=YELLOW_E, radius=0.05).move_to(self.coords_to_point(-1.508, 0.0))
    tang1_point_label = TextMobject('$x_1$', color=YELLOW).scale(0.75).move_to(self.coords_to_point(-1.508, 0.2))
    tang1_point_to_graph = self.get_vertical_line_to_graph(-1.508, graph).set_color(PURPLE)

    tang2 = TangentLine(graph, 0.3492, length=20)
    tang2.set_color(ORANGE)
    tang2_label = TextMobject('$y+2.08=9.63(x+1.51)$', color=ORANGE).move_to([0.0, -2.0, 0.0])
    tang2_point = Dot(color=GOLD_E).move_to(self.coords_to_point(-1.292, 0.0))
    tang2_point_small = Dot(color=GOLD_E, radius=0.05).move_to(self.coords_to_point(-1.292, 0.0))
    tang2_point_label = TextMobject('$x_2$', color=ORANGE).scale(0.75).move_to(self.coords_to_point(-1.292, 0.2))
    tang2_point_to_graph = self.get_vertical_line_to_graph(-1.292, graph).set_color(PURPLE)

    tang3 = TangentLine(graph, 0.3708, length=20)
    tang3.set_color(RED)
    tang3_label = TextMobject('$y+0.35=6.43(x+1.29)$', color=RED).move_to([0.0, -2.0, 0.0])
    tang3_point_small = Dot(color=RED_E, radius=0.05).move_to(self.coords_to_point(-1.237, 0.0))
    tang3_point_label = TextMobject('$x_3$', color=RED_C).scale(0.75).move_to(self.coords_to_point(-1.187, -0.2))


    x0A = TextMobject("$x_0$", color=BLUE).shift(0.25*DOWN)
    x0B = TextMobject("$=$", color=BLUE).shift(0.25*DOWN)
    x0C = TextMobject("$0.9$", color=BLUE).shift(0.25*DOWN)
    x0 = [x0A, x0B, x0C]

    x1A = TextMobject("$x_1$", color=YELLOW).shift(0.75*DOWN)
    x1B = TextMobject("$=$", color=YELLOW).shift(0.75*DOWN)
    x1C = TextMobject("$-1.508$", color=YELLOW).shift(0.75*DOWN)
    x1 = [x1A, x1B, x1C]

    x2A = TextMobject("$x_2$", color=ORANGE).shift(1.25*DOWN)
    x2B = TextMobject("$=$", color=ORANGE).shift(1.25*DOWN)
    x2C = TextMobject("$-1.292$", color=ORANGE).shift(1.25*DOWN)
    x2 = [x2A, x2B, x2C]

    x3A = TextMobject("$x_3$", color=RED).shift(1.75*DOWN)
    x3B = TextMobject("$=$", color=RED).shift(1.75*DOWN)
    x3C = TextMobject("$-1.237$", color=RED).shift(1.75*DOWN)
    x3 = [x3A, x3B, x3C]

    x0B.next_to(x0A,RIGHT)
    x0C.next_to(x0B,RIGHT)
    x1A.align_to(x0A,LEFT)
    x1B.align_to(x0B,LEFT)
    x1C.align_to(x0C,LEFT)
    x2A.align_to(x1A,LEFT)
    x2B.align_to(x1B,LEFT)
    x2C.align_to(x1C,LEFT)
    x3A.align_to(x2A,LEFT)
    x3B.align_to(x2B,LEFT)
    x3C.align_to(x2C,LEFT)

    xgroup = VGroup(x0A, x1A, x2A, x3A)
    xbraces = Brace(xgroup, LEFT)
    xbrace_text = xbraces.get_text("X-intercepts get closer to true root").scale(0.625)



    self.play(ShowCreation(x0_point), ShowCreation(x0_point_small), ShowCreation(x0_point_label))
    self.wait(1)
    self.play(ShowCreation(x0_point_to_graph))
    self.play(MoveAlongPath(x0_point, x0_point_to_graph))
    self.wait(1)
    self.play(Uncreate(x0_point_to_graph))
    self.wait(2.5)

    self.play(ShowCreation(tang1), Write(tang1_label), run_time=4.0)
    self.wait(1)
    self.play(ShowCreation(tang1_point), ShowCreation(tang1_point_small), ShowCreation(tang1_point_label))
    self.wait(1)
    self.play(ShowCreation(tang1_point_to_graph))
    self.play(MoveAlongPath(tang1_point, tang1_point_to_graph))
    self.wait(1)
    self.play(Uncreate(tang1), Uncreate(tang1_label), Uncreate(tang1_point_to_graph), Uncreate(x0_point), run_time=3.0)
    self.wait(2.5)

    self.play(ShowCreation(tang2), Write(tang2_label), run_time=4.0)
    self.wait(1)
    self.play(ShowCreation(tang2_point), ShowCreation(tang2_point_small), ShowCreation(tang2_point_label))
    self.wait(1)
    self.play(ShowCreation(tang2_point_to_graph))
    self.play(MoveAlongPath(tang2_point, tang2_point_to_graph))
    self.wait(1)
    self.play(Uncreate(tang2), Uncreate(tang2_label), Uncreate(tang2_point_to_graph), Uncreate(tang1_point), run_time=3.0)
    self.wait(2.5)

    self.play(ShowCreation(tang3), Write(tang3_label), run_time=4.0)
    self.wait(1)
    self.play(ShowCreation(tang3_point_small), ShowCreation(tang3_point_label))
    self.wait(1)
    self.play(Uncreate(tang3), Uncreate(tang3_label), Uncreate(tang2_point), run_time=3.0)
    self.wait(2.5)

    self.play(*[Write(x) for x in x0])
    self.play(*[Write(x) for x in x1])
    self.play(*[Write(x) for x in x2])
    self.play(*[Write(x) for x in x3])
    self.play(ShowCreation(xbraces), Write(xbrace_text))
    self.wait(3)




  def construct(self):
    x_label = TextMobject('$x$').move_to([6.75, 1.25, 0])
    y_label = TextMobject('$y$').move_to([4.50, 3.75, 0])
    origin_label = TextMobject('$O$').move_to([4.25, 0.75, 0])
    plane = NewNumberPlane()

    self.wait(2)
    self.play(ShowCreation(plane))
    self.setup_axes(animate=True)
    self.wait(1)
    self.play(Write(x_label), Write(y_label), Write(origin_label))    

    self.tangents()

   
    self.play(*[Uncreate(mob) for mob in self.mobjects], run_time=5.0)
    self.wait(1)


```

```
%%manim NewtonPART3
from manimlib.scene.scene import *
from manimlib.mobject.geometry import *
from manimlib.animation.creation import *
from manimlib.imports import *

class NewtonPART3(Scene):

  def construct(self):

    x3A = TextMobject("$x_3$", color=RED)
    x3B = TextMobject("$=$", color=RED).move_to([0.0, 2.75, 0.0])
    x3C = TextMobject("$-1.237$", color=RED)
    x3A.next_to(x3B,LEFT)
    x3C.next_to(x3B,RIGHT)
    x3 = [x3A, x3B, x3C]

    xrA = TextMobject("$x_r$", color=GREEN_SCREEN)
    xrB = TextMobject("$=$", color=GREEN_SCREEN).move_to([0.0, 1.75, 0.0])
    xrC = TextMobject("$-1.234$", color=GREEN_SCREEN)
    xrA.next_to(xrB,LEFT)
    xrC.next_to(xrB,RIGHT)
    xr = [xrA, xrB, xrC]

    explain1 = TextMobject('After just 3 iterations of Newton\'s method').set_color_by_gradient(BLUE, GREEN)
    explain2 = TextMobject('we can approximate the root to an error of just').set_color_by_gradient(BLUE, GREEN).shift(0.5*DOWN)
    explain3 = TextMobject('$0.003$, or $0.24\%$').set_color_by_gradient(BLUE, GREEN).shift(DOWN)


    self.play(*[Write(x) for x in x3])
    self.wait(1)
    self.play(*[Write(x) for x in xr])
    self.wait(1)
    self.play(Write(explain1))
    self.play(Write(explain2))
    self.wait(1)
    self.play(Write(explain3))
    self.wait(3)

    self.play(*[Uncreate(mob) for mob in self.mobjects], run_time=5.0)
    self.wait(1)
```
